# Student Performance Analysis

This notebook fixes the average calculation, creates additional student performance features, and visualizes the data using Matplotlib.

## 1. Average Marks Calculation

```python
marks = [45, 60, 72, 38, 90]

total = 0

for i in range(len(marks)):
    total += marks[i]

average = total / len(marks)

if average > 50:
    print("Class passed")
else:
    print("Class failed")

print("Average:", average)
```

## 2. Create Student DataFrame

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "student_id": [101, 102, 103, 104, 105, 106],
    "attendance_percent": [92, 67, 81, 45, 74, 88],
    "assignment_score": [18, 12, 15, 8, 14, 19],
    "quiz_score": [72, 48, 65, 30, 55, 80],
    "lab_completed": [True, False, True, False, True, True]
}

df = pd.DataFrame(data)
df
```

## 4. Feature Engineering

```python
df["total_score"] = df["assignment_score"] + df["quiz_score"]
df["performance_percent"] = (df["total_score"] / 120) * 100

df["attendance_status"] = df["attendance_percent"].apply(
    lambda value: "Good" if value >= 75 else "Low"
)

df["quiz_status"] = df["quiz_score"].apply(
    lambda value: "Pass" if value >= 50 else "Fail"
)

df["assignment_status"] = df["assignment_score"].apply(
    lambda value: "Good" if value >= 15 else "Needs Improvement"
)

df["eligible"] = (
    (df["attendance_percent"] >= 75) &
    (df["total_score"] >= 70) &
    (df["lab_completed"])
)

df
```

## 5. Eligible Students

```python
eligible_students = df[df["eligible"] == True]
eligible_students
```

## 6. Matplotlib Visualizations

```python
plt.figure(figsize=(8, 5))
plt.bar(df["student_id"], df["attendance_percent"], color="skyblue")
plt.axhline(y=75, color="red", linestyle="--", label="Required Attendance")
plt.xlabel("Student ID")
plt.ylabel("Attendance Percentage")
plt.title("Student Attendance Percentage")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.bar(df["student_id"], df["assignment_score"], label="Assignment Score", color="orange")
plt.bar(
    df["student_id"],
    df["quiz_score"],
    bottom=df["assignment_score"],
    label="Quiz Score",
    color="green"
)
plt.xlabel("Student ID")
plt.ylabel("Score")
plt.title("Assignment and Quiz Scores")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(df["student_id"], df["total_score"], marker="o", color="purple")
plt.xlabel("Student ID")
plt.ylabel("Total Score")
plt.title("Total Score of Students")
plt.grid(True)
plt.show()
```

```python
eligible_counts = df["eligible"].map({True: "Eligible", False: "Not Eligible"}).value_counts()

plt.figure(figsize=(6, 6))
plt.pie(
    eligible_counts,
    labels=eligible_counts.index,
    autopct="%1.1f%%",
    colors=["lightgreen", "lightcoral"]
)
plt.title("Eligibility Distribution")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df["attendance_percent"], df["total_score"], color="blue")

for index, row in df.iterrows():
    plt.text(
        row["attendance_percent"] + 0.5,
        row["total_score"],
        row["student_id"]
    )

plt.xlabel("Attendance Percentage")
plt.ylabel("Total Score")
plt.title("Attendance vs Total Score")
plt.grid(True)
plt.show()
```
