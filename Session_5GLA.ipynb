```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url="https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv"

df=pd.read_csv(url)


print(df.head())
print(df.shape)
print(df.columns)
print(df.info())
```

```python
print("Minimum Year:", df["year"].min())
print("Maximum Year:", df["year"].max())
print("range", df["year"].min(),"-",df["year"].max())
num_countries = df["country"].nunique()
print("Number of Countries:", num_countries)
num_continents = df["continent"].nunique()
print("Number of Continents:", num_continents)
print(df["continent"].unique())
# Total records
print(df.shape[0])

# Total columns
print(df.shape[1])

# Average life expectancy
print(df["lifeExp"].mean())

# Highest GDP per capita
print(df["gdpPercap"].max())

# Country count by continent
print(df.groupby("continent")["country"].nunique())
```

```python
import matplotlib.pyplot as plt

life_exp = df.groupby("continent")["lifeExp"].mean()

print(life_exp)

plt.figure(figsize=(8,5))
plt.bar(life_exp.index, life_exp.values)

plt.title("Average Life Expectancy by Continent")
plt.xlabel("Continent")
plt.ylabel("Average Life Expectancy")

plt.show()
```

```python
import matplotlib.pyplot as plt


life_exp = df.groupby("continent")["lifeExp"].sum()

plt.bar(life_exp.index, life_exp.values)
plt.xlabel("Continent")
plt.ylabel("Total Life Expectancy")
plt.title("Total Life Expectancy by Continent")
plt.show()
```

```python
latest_year = df["year"].max()
latest_df = df[df["year"] == latest_year].copy()

print("Latest year:", latest_year)
print(latest_df.head())
print(latest_df.shape)
```

```python
print(df["continent"].value_counts())

continent_summary = df.groupby("continent")[[
    "lifeExp",
    "gdpPercap",
    "pop"
]].mean()

print(continent_summary)
```

```python
# Select countries
countries = ["India", "China", "United States"]

# Filter data
filtered_df = df[df["country"].isin(countries)]

# Plot
plt.figure(figsize=(10, 8))

for country in countries:
    country_data = filtered_df[filtered_df["country"] == country]
    plt.plot(country_data["year"],
             country_data["lifeExp"],
             marker='o',
             label=country)

plt.title("Life Expectancy Over Time")
plt.xlabel("Year")
plt.ylabel("Life Expectancy")
plt.legend()
plt.grid(True)

plt.show()
```

```python
# Scatter plot
plt.figure(figsize=(8,6))
plt.scatter(df["gdpPercap"], df["lifeExp"])

plt.title("GDP per Capita vs Life Expectancy")
plt.xlabel("GDP per Capita")
plt.ylabel("Life Expectancy")

plt.show()
```
