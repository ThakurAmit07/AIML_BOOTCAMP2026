```python
students= {
    "Amit": {"study_hours": 5, "attendance": 80, "marks": 85},
    "Anshul":  {"study_hours": 3, "attendance": 70, "marks": 45},
    "Ansh":   {"study_hours": 6, "attendance": 90, "marks": 78},
    "Ashutosh":{"study_hours": 2, "attendance": 60, "marks": 15},
    "Sahaj":   {"study_hours": 4, "attendance": 76, "marks": 35}
}

name = input("Enter the student name: ")

if name in students:
    data = students[name]
    marks = data["marks"]
    attendance = data["attendance"]
    study_hours = data["study_hours"]

    print(f"{name}")

    if marks > 75:
        print("Performance: Strong")
    elif marks >= 40:
        print("Performance: Needs improvement")
    elif marks < 20:
        print("Performance: Fail, repeat class")
    else:
        print("Performance: Poor, requires immediate attention")

    if attendance < 75:
        print("Warning: Low attendance")

    if study_hours < 4:
        print("Advice: Needs more study time")

else:
    print("Error: Student not found in the directory.")
```

```python
# Student Starter Code
transactions = [
    {"transaction_id": "TX101", "amount": 6200, "method": "Credit Card"},
    {"transaction_id": "TX102", "amount": 5500, "method": "Gift Card"},
    {"transaction_id": "TX103", "amount": 1200, "method": "Gift Card"},
    {"transaction_id": "TX104", "amount": 8000, "method": "Gift Card"},
    {"transaction_id": "TX105", "amount": 450, "method": "PayPal"}
]

flagged_transactions = []



for i in transactions:
    if i["amount"] > 5000 and i["method"] == "Gift Card":
        flagged_transactions.append(i["transaction_id"])

print(flagged_transactions)
```
